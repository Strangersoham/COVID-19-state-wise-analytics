## Introduction

 COVID-19 is an infectious disease caused by the Corona Virus, biologically known as severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). The disease was first identified in Wuhan, the capital of China's Hubei province in December 2019 and has spread all over the world since then.

##  Finding Indian states which are at high risk of encountering COVID 3rd Wave

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
covid = pd.read_csv('Latest Covid-19 India Status.csv')
```

```python
covid
```

```python
covid.columns = ['state','total_cases','active','discharged','death','active_ratio','discharged_ratio','death_ratio']
```

```python
covid
```

```python

plt.figure(figsize=(20,10))
plt.plot(covid.state,covid.death)
plt.grid(True)
plt.title('Data of  deaths due to Covid 19')
plt.xlabel('state')
plt.ylabel('No. of Deaths')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(covid.state,covid.active,color= "green")
plt.grid(True)
plt.title('Data of active cases of Covid 19')
plt.xlabel('state')
plt.ylabel('No. of Deaths')
plt.xticks(rotation = 90)
plt.show()
```

```python
total_case_sort = covid.sort_values(by='total_cases',ascending = False)
plt.figure(figsize= (20,10))
sns.barplot(data = total_case_sort,x ='state',y='total_cases')
plt.xticks(rotation = 90)
plt.show()
```

```python
sorted_data = covid.sort_values(by='death_ratio', ascending=False).head(10)
plt.bar(sorted_data['state'],sorted_data['death_ratio'])

plt.xlabel('state')
plt.ylabel('Percentage(death ratio)')
plt.title('highest number of deaths')
plt.xticks(rotation=90)
plt.show()
```

### From our visualizations we can say that Kerala, Karnataka, Tamil Nadu and Maharastra are at high risks of having a COVID-19 3rd wave.
